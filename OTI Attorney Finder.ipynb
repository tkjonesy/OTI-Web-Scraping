# CSK Legal

```python
driver.get('https://csklegal.com/attorneys/')
```

```python
data = []

container = driver.find_element(By.CLASS_NAME, 'attorny_filter_data')
att_list = container.find_elements(By.TAG_NAME, 'section')

for person in att_list:
    attorney = {}
    try:
        name = person.find_element(By.CLASS_NAME, 'main-name').text
        position = person.find_element(By.CLASS_NAME, 'post-tag09').text
        location = person.find_element(By.CLASS_NAME, 'place-tag09').text
        contact_info = person.find_element(By.CLASS_NAME, 'info-tag09')
        contacts = contact_info.find_elements(By.TAG_NAME, 'h4')
        phone = contacts[0].text
        email = contacts[1].text

        attorney['Name'] = name
        attorney['Email'] = email
        attorney['Phone #'] = phone
        attorney['Position'] = position
        attorney['Location'] = location
    except:
        continue
    
    data.append(attorney)
    
```

# Resnick & Louis

```python
driver.get('https://www.rlattorneys.com/attorneys/')
```

```python
fl_locations = ['Fort Lauderdale', 'Jacksonville', 'Miami', 'Orlando', 'Sarasota', 'Tampa']
```

```python
data = []
```

```python
container = driver.find_element(By.CLASS_NAME, 'memberBoxrow')
results = container.find_elements(By.XPATH,"//*[contains(@id, 'not')]")
for result in results:
    person = {}
    
    name_split = result.text.split(',')
    name = name_split[0].strip()
    position = name_split[1].strip()
    
    profile = result.find_element(By.TAG_NAME, 'a')
    profile.click()
    
    driver.implicitly_wait(random.uniform(1.0, 2.0))
    
    main = driver.find_element(By.CLASS_NAME, 'locationpglft')
    location_links = main.find_elements(By.XPATH, "//a[contains(@href, 'https://www.rlattorneys.com/location/')]")
    
    for x in location_links:
        if x.text != '':
            location = x.text
            break
            
    if location not in fl_locations:
        print(f'{name} works in {location}')
        driver.back()
        continue

    email = main.find_element(By.PARTIAL_LINK_TEXT, '@rlattorneys.com').text
    phone = main.find_element(By.XPATH, "//a[contains(@href, 'tel:')]").text
    
    person['Name'] = name
    person['Email'] = email
    person['Phone #'] = phone
    person['Location'] = location
    person['Position'] = position
    
    data.append(person)
    
    driver.back()
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
df1.to_excel('Resnick.Louis.xlsx', index=False)
```

# Fowler White Burnett

```python
driver.get('https://fowler-white.com/Attorneys?pa=12')
```

```python
data = []
```

```python
results = driver.find_element(By.CLASS_NAME, 'attorneyResults')
people = results.find_elements(By.CLASS_NAME, 'news-item')

for person in people:
    attorney = {}
    summary = person.find_element(By.CLASS_NAME, 'at-summary')
    sum_text = summary.text
    name_link = summary.find_element(By.TAG_NAME, 'a')
    name = name_link.find_element(By.TAG_NAME, 'strong').text
    position = summary.find_element(By.TAG_NAME, 'em').text
    
    if position == 'RETIRED':
        print(f'{name} is Retired')
        continue
    
    area = summary.find_element(By.TAG_NAME, 'span').text
    
    try:
        phone = re.findall(r'\d{3}.\d{3}.\d{4}', sum_text)[0]
    except:
        print(f'{name} cannot find phone number')
        phone = 'Not Found'
    
    try:
        email = summary.find_element(By.PARTIAL_LINK_TEXT, '@fowler-white.com').text
    except:
        print(f'{name} cannot find email')
        email = 'Not Found'

    attorney['Name'] = name
    attorney['Email'] = email
    attorney['Phone #'] = phone
    attorney['Location'] = area
    attorney['Position'] = position
    
    data.append(attorney)
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
df1.to_excel('Fowler.White.xlsx', index=False)
```

# Martindale

```python
driver.get('https://www.martindale.com/')
```

```python
data = []
```

```python
resultdiv = driver.find_element(By.CLASS_NAME, 'results__result-row')
resultslist = resultdiv.find_elements(By.CLASS_NAME, 'card--firm')

for result in resultslist:
    company = {}
    card_details = result.find_element(By.CLASS_NAME, 'card__content--details')
    name = card_details.find_element(By.CLASS_NAME, 'detail_title').text.replace('Open for Business', '')
    name = name.strip()
    area = card_details.find_element(By.CLASS_NAME, 'detail_location').text.replace('Serving ', '')
    
    try:
        contact_card = result.find_element(By.CLASS_NAME, 'card__content--contact')
        
        try:
            phone_raw = contact_card.find_element(By.CLASS_NAME, 'webstats-phone-click')
            
            if 'VIEW PHONE' in phone_raw.text:
                phone = phone_raw.get_attribute('data-phone')
                
            else:
                phone = phone_raw.get_attribute('href').replace('tel:', '')
            
        except:
            print(f'{name} does not have a phone number')
            phone = None
            
        try:
            website = contact_card.find_element(By.CLASS_NAME, 'webstats-website-click').get_attribute('href')
        
        except:
            print(f'{name} does not have a website')
            website = None
    except:
        print(f'{name} does not have contact info')
        phone = None
        website = None
        
    company['Name'] = name
    company['Website'] = website
    company['Phone'] = phone
    company['Area'] = area
    
    data.append(company)
```

```python
df = pd.DataFrame(data)
df1 = df.drop_duplicates()
df2 = df1.drop_duplicates(subset=df.columns.difference(['Area']))
df1.to_excel('Raw.Law.Offices.FL.xlsx', index=False)
df2.to_excel('Clean.Law.Offices.FL.xlsx', index=False)
df1
```
